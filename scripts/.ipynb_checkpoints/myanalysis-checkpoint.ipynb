### Introduction 

Reproducibility is a common issue in neuroimaging and neuroscience. One of the contributing factors to reproducibility failures is **p-hacking**, which hypotheses are determined after data wrangling and discovery of statistical significance. This analysis simulates the process of p-hacking using brain size data. A random variable is first generated to yield significant p-values in a linear regression model. The same model is then re-run using a second random variable to show the reproducibility failure. Visualizations are produced along the way to aid the analysis. 

### Reading data 

We start by importing essential libraries and setting the path of the brain size csv file. Raw data are read as pandas dataframe with the following settings: ";" as the separator, "." as missing values, first row as header, and first column as index. We then drop rows with missing data and display the first five rows of the dataframe. 

```python
# import libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing 
import statsmodels.formula.api as smf

# set raw data path 
csv_file = "../data/brainsize.csv"

# read data from csv 
brainsize_df = pd.read_csv(csv_file, sep=";", na_values=".", header=0, index_col=0)

# drop rows with N/A values 
brainsize_df = brainsize_df.dropna()

# preview dataframe 
print(brainsize_df.head())
```

### Data exploration 

Summary statistics are printed for categorical and numeric variables. Among the 38 participants, males and females are almost equally distributed. FSIQ, VIQ and PIQ have very close mean and standard deviations, while weight, height and MRI_count have more different ranges. 

```python
# view summary statistics 
print(brainsize_df.describe())
print("\n")
print(brainsize_df["Gender"].value_counts())
```

We use `seaborn.distplot` to plot the distribution of continuous variables. While FSIQ and Weight have two modals, Height and MRI_count have one modal instead. The distribution of VIQ and PIQ are more uniform. 

```python
# create subplots 
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8,4))

# plot distributions of numeric columns 
for i, col in enumerate(brainsize_df.columns[1:]):
    sns.distplot(brainsize_df[col], ax=axes[i//3, i%3])

# adjust layouts
plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)
```

### Generating random variables 

We use `numpy.random` to generate random variables partY and partY2. The means and standard deviations of the random variables are set close to FSIQ, VIQ and PIQ. Both partY and partY2 have normal distributions. 

```python
# generate random variables using normal distribution 
np.random.seed(530) # set seed 
mu, sigma = 110, 23 # mean, std. dev.
partY = np.random.normal(mu, sigma, 38)
brainsize_df["partY"] = partY # attach partY to brainsize_df

np.random.seed(61) # change seed 
partY2 = np.random.normal(mu, sigma, 38)
brainsize_df["partY2"] = partY2 # attach partY2 to brainsize_df

print(brainsize_df.head())
```

### Initial modeling 

We first implement Ordinary Least Squares (OLS) regression models using `statsmodels`. After some experiments, PIQ is chosen to be the independent variable against partY bevause of its significant p-value (0.021). However, after replacing the dependent variable with partY2, the p-value of PIQ (0.379) lose its significance. 

```python
# fit linear models using Ordinary Least Squares
model = ols('partY ~ PIQ', brainsize_df).fit()
print('partY ~ PIQ\nModel: OLS\nData: Original brainsize')
print(model.pvalues)

model = ols('partY2 ~ PIQ', brainsize_df).fit()
print('\npartY2 ~ PIQ\nModel: OLS\nData: Original brainsize')
print(model.pvalues)
```

### Improving p-values 

We try reducing the p-values through data normalization and Generalized Linear Model (GLM). Using `MinMaxScaler` from `sklearn.preprocessing`, we re-scale all continuous variables to the [0, 1] range. As a result, the p-value of the intercept (0.001) decreased, but the p-value of PIQ (0.021) remain the same. We then implement the GLM model from `statsmodels` on normalized brain size data. This final model helps decrease both p-values. Therefore, we re-run the GLM model with partY2. Although both p-values are better than those from the OLS model, the p-value of PIQ (0.373) is still too large. 

```python
# data normalization using sklearn.preprocessing 
x = brainsize_df.iloc[:,1:].values # returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
normalized_df = pd.DataFrame(x_scaled)
normalized_df.columns = brainsize_df.columns[1:] # update column names 

# fit an OLS model using normalized data 
model = ols('partY ~ PIQ', normalized_df).fit()
print('partY ~ PIQ\nModel: OLS\nData: Normalized brainsize')
print(model.pvalues)

# fit Generalized Linear Models using normalized data
model = smf.glm('partY ~ PIQ', normalized_df).fit()
print('\npartY ~ PIQ\nModel: GLM\nData: Normalized brainsize')
print(model.pvalues)
print(model.summary())

model = smf.glm('partY2 ~ PIQ', normalized_df).fit()
print('\npartY2 ~ PIQ\nModel: GLM\nData: Normalized brainsize')
print(model.pvalues)
print(model.summary())
```

### Conclusion 

The analysis is concluded by the visualizations of the final models. Using `seaborn.regplot`, it can be seen that partY2 and PIQ has a weaker correlation than partY and PIQ because of its flatter regression line. It is not a surprise that the significant p-values are non-reproducible on partY2 as both partY and partY2 are random variables. In short, p-hacking needs to be prevented to increase reproducibility of scientific studies. Further experiments can be conducted on categorical variables and polynomial regression models, but similar results can be expected with additional attempts of p-hacking. 

```python
# create subplots and set figure size 
fig, ax =plt.subplots(1,2, figsize=(8,4))

# plot data and linear regression models 
sns.regplot(x=normalized_df['PIQ'], y=normalized_df['partY'], ax=ax[0])
sns.regplot(x=normalized_df['PIQ'], y=normalized_df['partY2'], ax=ax[1], color="orange")

# adjust layouts
plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)
```
